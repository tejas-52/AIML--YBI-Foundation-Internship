```python
#Step1: import library
import pandas as pd
```

```python
#Step2: import data
house= pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Boston.csv')
house
```

```python
#Step3: define your target (y) and features (x)
```

```python
house.columns
```

```python
y= house['MEDV']

x=house[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX',
       'PTRATIO', 'B', 'LSTAT']]
```

```python
x.shape
```

```python
y.shape
```

```python
#Train test split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.75,random_state=2529)
x_train
```

```python
x_test.shape,x_train.shape,y_test.shape,y_train.shape
```

```python
#Step5: Select model
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
#Step6: Train the model(fit model)
model.fit(x_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
#step7:  prediction
y_prdict= model.predict(x_test)
y_prdict
```

```python
#Step8: accuracy
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mean_absolute_percentage_error(y_test,y_prdict)
```
